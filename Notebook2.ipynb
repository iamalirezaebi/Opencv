### In this cell we imported the neccessary packages to work with in this notebook three lines at the end were got from stackoverflow website in order to solve systemexit2 error in the followin cell "args = parser.parse_args()"

```python
from __future__ import print_function
import cv2 as cv
import argparse
import math
import sys
sys.argv=['']
del sys
```

```python
vehicles = 0
vehicles_per_minute = 0
```

### this cell is for input algorithm and input video the resource is open cv official website

```python
parser = argparse.ArgumentParser(description='This program shows how to use background subtraction methods provided by \
                                              OpenCV. You can process both videos and images.')
parser.add_argument('--input', type=str, help='Path to a video or a sequence of image.', default='Traffic_Laramie_2.mp4')
parser.add_argument('--algo', type=str, help='Background subtraction method (KNN, MOG2).', default='MOG2')
args = parser.parse_args()
```

### this cell initializes our approach to deal with in this execise

```python
if args.algo == 'MOG2':
    backSub = cv.createBackgroundSubtractorMOG2(history = 100,varThreshold= 5)
else:
    backSub = cv.createBackgroundSubtractorKNN()
capture = cv.VideoCapture(cv.samples.findFileOrKeep(args.input))
if not capture.isOpened():
    print('Unable to open: ' + args.input)
    exit(0)
```

### Video FPS Retrieval and Frame Count Initialization

```python
fps = capture.get(cv.CAP_PROP_FPS)
frame_count = 0
```

```python
while True:
    capture.set(3, 1920)
    capture.set(4, 1080)
    ret, frame = capture.read()
    frame_count += 1
    time_elapsed = frame_count / fps
    height, width, _ = frame.shape
    
    # Extract Region of interest
    roi = frame[340: 720,: 1080]
    
    # 1. Object Detection
    fgMask = backSub.apply(roi)
    _, fgMask = cv.threshold(fgMask, 254, 255, cv.THRESH_BINARY)
    contours, _ = cv.findContours(fgMask, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
    matches = []
    num = 0
    def get_centrolid(x, y, w, h):
        x1 = int(w /2)
        y1 = int(h / 2)
 
        cx = x + x1
        cy = y + y1
        return cx, cy
    for cnt in contours:
        # Calculate area and removedetections small elements
        area = cv.contourArea(cnt)

        if area > 4000:
            #cv.drawContours(roi, [cnt], -1, (0, 255, 0), 2)
            x, y, w, h = cv.boundingRect(cnt)
            
            cv.rectangle(roi, (x-10,y-10),(x+w,y+h),(0,255,0),3)
            cv.line(roi, (0,550), (550,0), (0, 255, 0), 3)
            centrolid = get_centrolid(x, y, w, h)
            matches.append(centrolid)
            cv.circle(roi, centrolid, 5, (0, 255, 0), -1)
            cx, cy = get_centrolid(x, y, w, h)
            for (cx,cy) in matches:
                if 165.9<cx<170 and cy<90:
                    vehicles = vehicles +1
                    matches.remove((cx, cy))
                    
    if time_elapsed >= 60:
        vehicles_per_minute = vehicles
        frame_count = 0




    # Display the total count of cars moving from downtown to the city center
    cv.putText(frame, f"Cars moving to center: {vehicles}", (10, 50), cv.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)
    cv.putText(frame, f"Cars per minute: {vehicles_per_minute}", (10, 90), cv.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)

    cv.imshow('Frame', frame)
    cv.imshow('FG Mask', fgMask)
    keyboard = cv.waitKey(1)
    if keyboard == 'q' or keyboard == 27:
        break
```
